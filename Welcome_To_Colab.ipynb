```
import pandas as pd
data =pd.read_csv("student_feedback.csv")
data.head()
```

```
data.tail(10)
```

```
data=data.drop(columns='Unnamed: 0',errors='ignore')
data
```

```
data.info()
data.describe(include='all')
```

```
import matplotlib.pyplot as plt
import seaborn as sns
df=data.drop(columns='Student ID',errors='ignore')
avg_rating=df.mean()
plt.figure(figsize=(10,5))
sns.barplot(x=avg_rating.values,y=avg_rating.index)
plt.title('avg feeedback rating')
plt.xlabel('values')
plt.ylabel('categerie')
plt.show
```

```
import matplotlib.pyplot as plt
import seaborn as sns
df=data.drop(columns='Student ID',errors='ignore')
avg_rating=df.mean()
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.title('avg feeedback rating')
plt.xlabel('values')
plt.ylabel('categerie')
plt.show
```

```
import matplotlib.pyplot as plt
import seaborn as sns
df=data.drop(columns='Student ID',errors='ignore')
avg_rating=df.mean()
plt.figure(figsize=(10,5))
sns.boxplot(data=df, orient='h', palette='Set1')
plt.title('avg feeedback rating')
plt.xlabel('values')
plt.ylabel('categerie')
plt.show
```

```
import matplotlib.pyplot as plt
import seaborn as sns
df=data.drop(columns='Student ID',errors='ignore')
avg_rating=df.mean()
plt.figure(figsize=(10,5))
sns.histplot(data=df,x='Degree of difficulty of assignments',bins=10)
plt.title('avg feeedback rating')
plt.xlabel('values')
plt.ylabel('categerie')
plt.show
```

```
import matplotlib.pyplot as plt
import seaborn as sns
df=data.drop(columns='Student ID',errors='ignore')
avg_rating=df.mean()
plt.figure(figsize=(10,5))
sns.kdeplot(data=df['Course recommendation based on relevance'],fill=True)
plt.title('avg feeedback rating')
plt.xlabel('values')
plt.ylabel('categerie')
plt.show
```
